```python
import pandas as pd
```

SQL Querys

```python
DROP TABLE IF EXISTS meteorite;

CREATE TABLE meteorite (
	index INT PRIMARY KEY NOT NULL,
	name VARCHAR(100) NOT NULL,
	classification VARCHAR(250) NOT NULL,
	mass DOUBLE PRECISION NOT NULL,
	fall VARCHAR(50) NOT NULL,
	year INT NOT NULL,
	latitude DOUBLE PRECISION NOT NULL,
	longitude DOUBLE PRECISION NOT NULL
);

SELECT * FROM meteorite;
```

```python
meteor_df = pd.read_csv("Meteorite_Landings.csv")
```

```python
meteor_df
```

```python
# Drop rows with missing 'year' data

# meteor_df.dropna(subset=['year'], inplace=True)

meteor_df.dropna(inplace=True)
meteor_df
```

```python
# Verify NaN are removed
meteor_df['year'].unique()
```

```python
# Drop 0,0 coordinates
# df.drop(df[df['Fee'] >= 24000].index, inplace = True)
meteor_df.drop(meteor_df[meteor_df['reclat'] == 0.0].index, inplace=True)
```

```python
# Confirm drop
meteor_df
```

```python
# Convert year data type to integer
meteor_df['year'] = meteor_df['year'].astype('int')
meteor_df.info()
```

```python
# Drop un-needed columns
cleaned_meteor_df = meteor_df.drop(columns=['id', 'GeoLocation', 'nametype'])
```

```python
# Rename columns
cleaned_meteor_df = cleaned_meteor_df.rename(columns={'reclat':'latitude', 'reclong':'longitude', 'recclass':'classification', ''})
```

```python
# Reset index positions
cleaned_meteor_df.reset_index(drop=True, inplace=True)
```

```python
# Correct invalid date
cleaned_meteor_df.iloc[30411, 5] = 2010
```

```python
cleaned_meteor_df
```

```python
cleaned_meteor_df.to_csv('cleaned_meteorite_landing.csv')
```
